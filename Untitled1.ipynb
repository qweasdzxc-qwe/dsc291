```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize, brentq
import scipy.optimize as opt
```

```python
dfA = pd.read_csv("DetectorA.csv")
dfB = pd.read_csv("DetectorB.csv")
dfC = pd.read_csv("DetectorC.csv")
dfTarget = pd.read_csv("DetectorTarget.csv")

print(f"Events in Detector A: {len(dfA)}")
print(f"Events in Detector B: {len(dfB)}")
print(f"Events in Detector C: {len(dfC)}")
print(f"Events in Detector Target: {len(dfTarget)}")

energy_min, energy_max, bin_width = 1000, 3500, 1
bins = np.arange(energy_min, energy_max + bin_width, bin_width)

counts_A, _ = np.histogram(dfA["energy"], bins=bins)
counts_B, _ = np.histogram(dfB["energy"], bins=bins)
counts_C, _ = np.histogram(dfC["energy"], bins=bins)
counts_Target, _ = np.histogram(dfTarget["energy"], bins=bins)

plt.figure(figsize=(8,6))
plt.hist(dfA["energy"], bins=bins, histtype='step', label='Detector A')
plt.hist(dfB["energy"], bins=bins, histtype='step', label='Detector B')
plt.hist(dfC["energy"], bins=bins, histtype='step', label='Detector C')
plt.hist(dfTarget["energy"], bins=bins, histtype='step', label='Detector Target')
plt.xlabel("Energy (keV)")
plt.ylabel("Counts per 1 keV bin")
plt.title("Energy Spectra of Detectors A, B, C, and Target")
plt.yscale("log")
plt.legend()
plt.show()
```

```python
peak_low_B, peak_high_B = 2102, 2104
peak_window_B = (dfB["energy"] >= peak_low_B) & (dfB["energy"] <= peak_high_B)

# Example sidebands: 20962101 keV, 21052110 keV
sb_left_low_B, sb_left_high_B = 2096, 2101
sb_right_low_B, sb_right_high_B = 2105, 2110

sb_left_B = (dfB["energy"] >= sb_left_low_B) & (dfB["energy"] < sb_left_high_B)
sb_right_B = (dfB["energy"] >= sb_right_low_B) & (dfB["energy"] < sb_right_high_B)

peakB_events = dfB[peak_window_B]
sidebandB_events = pd.concat([dfB[sb_left_B], dfB[sb_right_B]])

# 2) Count how many total events are in peak vs. sidebands
peak_counts_B = len(peakB_events)
sideband_counts_B = len(sidebandB_events)

# 3) Calculate widths (keV) for peak and sidebands
peak_width_B = peak_high_B - peak_low_B            # 2 keV
sideband_width_B = (sb_left_high_B - sb_left_low_B) + \
                   (sb_right_high_B - sb_right_low_B)  # 5 + 5 = 10 keV

# 4) Estimate background under the peak
bg_rate_B = sideband_counts_B / sideband_width_B
bg_in_peak_B = bg_rate_B * peak_width_B
signal_peak_B = peak_counts_B - bg_in_peak_B  # net "peak" events after bkg subtraction

print("Raw peak counts (2102-2104):", peak_counts_B)
print("Sideband counts (2096-2101, 2105-2110):", sideband_counts_B)
print("Background estimate under peak:", bg_in_peak_B)
print("=> Net background peak (2103 keV):", signal_peak_B)

# 5) Choose a CNN threshold
threshold_B = 0.10

# 6) How many events in the peak pass this threshold?
scores_peak_B = peakB_events["cnn_score"].values
n_peak_pass_B = np.sum(scores_peak_B >= threshold_B)

# 7) How many sideband events pass the threshold?
scores_sideband_B = sidebandB_events["cnn_score"].values
n_sideband_pass_B = np.sum(scores_sideband_B >= threshold_B)

bg_pass_rate_B = n_sideband_pass_B / sideband_width_B
bg_in_peak_pass_B = bg_pass_rate_B * peak_width_B
signal_pass_B = n_peak_pass_B - bg_in_peak_pass_B

# 8) FPR = fraction of this background peak that passes
FPR_B = 0
if signal_peak_B > 0:
    FPR_B = signal_pass_B / signal_peak_B

print("Threshold:", threshold_B)
print("Peak events passing cut:", n_peak_pass_B)
print("Sideband pass scaled to peak:", bg_in_peak_pass_B)
print("=> Net background passing cut:", signal_pass_B)
print(f"False Positive Rate (baseline-sub): {FPR_B*100:.2f}%")

# (Optional) Plot CNN score distribution for 2103 keV peak
plt.figure(figsize=(6,4))
plt.hist(scores_peak_B, bins=20, range=(0,1), color='orange', edgecolor='black')
plt.axvline(threshold_B, color='red', linestyle='--', 
            label=f"Threshold = {threshold_B}")
plt.title("CNN Score (2103 keV Peak) with Baseline Subtraction")
plt.xlabel("CNN Score")
plt.ylabel("Number of Events")
plt.legend()
plt.show()
```

```python
peak_low, peak_high = 1591, 1593
peak_window = (dfA["energy"] >= peak_low) & (dfA["energy"] <= peak_high)

sideband_left_low, sideband_left_high = 1585, 1590
sideband_right_low, sideband_right_high = 1595, 1600

sb_left = (dfA["energy"] >= sideband_left_low) & (dfA["energy"] < sideband_left_high)
sb_right = (dfA["energy"] >= sideband_right_low) & (dfA["energy"] < sideband_right_high)

peakA_events = dfA[peak_window]
sidebandA_events = pd.concat([dfA[sb_left], dfA[sb_right]])

# 2) Count how many total events are in peak vs sidebands
peak_counts = len(peakA_events)
sideband_counts = len(sidebandA_events)

# 3) Calculate widths (keV) for peak and sidebands
peak_width = peak_high - peak_low            # 2 keV
sideband_width = (sideband_left_high - sideband_left_low) + \
                 (sideband_right_high - sideband_right_low)  # 5 + 5 = 10 keV

# 4) Estimate background under the peak
bg_rate = sideband_counts / sideband_width
bg_in_peak = bg_rate * peak_width
signal_est = peak_counts - bg_in_peak

print("Raw peak counts (1591-1593):", peak_counts)
print("Sideband counts (1585-1590, 1595-1600):", sideband_counts)
print("Background estimate under peak:", bg_in_peak)
print("=> Net signal in 1592 keV peak:", signal_est)

# 5) Choose CNN threshold
threshold = threshold_B

# 6) How many events in the peak pass this threshold?
scores_peak = peakA_events["cnn_score"].values
n_peak_pass = np.sum(scores_peak >= threshold)

# 7) How many sideband events pass the threshold?
scores_sideband = sidebandA_events["cnn_score"].values
n_sideband_pass = np.sum(scores_sideband >= threshold)
bg_pass_rate = n_sideband_pass / sideband_width
bg_in_peak_pass = bg_pass_rate * peak_width
signal_pass_est = n_peak_pass - bg_in_peak_pass

# 8) Compute TPR = fraction of signal that passes
TPR = 0
if signal_est > 0:
    TPR = signal_pass_est / signal_est

print("Threshold:", threshold)
print("Peak events passing cut:", n_peak_pass)
print("Sideband pass scaled to peak:", bg_in_peak_pass)
print("=> True signal passing cut:", signal_pass_est)
print("True Positive Rate (with baseline subtraction):", f"{TPR*100:.2f}%")

# (Optional) Plot the CNN score distribution in the peak
plt.figure(figsize=(6,4))
plt.hist(scores_peak, bins=20, range=(0,1), color='skyblue', edgecolor='black')
plt.axvline(threshold, color='red', linestyle='--', label=f"Threshold={threshold}")
plt.title("CNN Score (1592 keV) with Baseline Subtraction")
plt.xlabel("CNN Score")
plt.ylabel("Number of Events")
plt.legend()
plt.show()
```

```python
import scipy.stats as stats


nldbd_mean = 2039.0 
nldbd_sigma = 1.0 

energy_bins = bins 
cdf_vals = stats.norm.cdf(energy_bins, loc=nldbd_mean, scale=nldbd_sigma)
nldbd_pdf = cdf_vals[1:] - cdf_vals[:-1] 
nldbd_pdf = nldbd_pdf / nldbd_pdf.sum() 

print(f"Sum of NLDBD PDF values: {nldbd_pdf.sum():.3f}")
peak_bin = np.argmax(nldbd_pdf)
print(f"NLDBD PDF peak at bin center ~{peak_bin + 0.5} keV with PDF value {nldbd_pdf[peak_bin]:.4f}")
```

```python
cut_thr = threshold

dfA_cut = dfA[dfA["cnn_score"] >= cut_thr]
dfB_cut = dfB[dfB["cnn_score"] >= cut_thr]
dfC_cut = dfC[dfC["cnn_score"] >= cut_thr]
dfTarget_cut = dfTarget[dfTarget["cnn_score"] >= cut_thr]

for name, original, cut_df in [("A", dfA, dfA_cut), ("B", dfB, dfB_cut), ("C", dfC, dfC_cut), ("Target", dfTarget, dfTarget_cut)]:
    frac = len(cut_df) / len(original) * 100
    print(f"Detector {name}: {len(cut_df)} events remain after cut ({frac:.1f}% of original)")

countsA_cut, _ = np.histogram(dfA_cut["energy"], bins=bins)
countsB_cut, _ = np.histogram(dfB_cut["energy"], bins=bins)
countsC_cut, _ = np.histogram(dfC_cut["energy"], bins=bins)
countsTarget_cut, _ = np.histogram(dfTarget_cut["energy"], bins=bins)

pdfA = countsA_cut / countsA_cut.sum() if countsA_cut.sum() > 0 else countsA_cut
pdfB = countsB_cut / countsB_cut.sum() if countsB_cut.sum() > 0 else countsB_cut
pdfC = countsC_cut / countsC_cut.sum() if countsC_cut.sum() > 0 else countsC_cut


plt.figure(figsize=(8,6))
plt.hist(dfA_cut["energy"], bins=bins, histtype='step', label='Detector A (cut)')
plt.hist(dfB_cut["energy"], bins=bins, histtype='step', label='Detector B (cut)')
plt.hist(dfC_cut["energy"], bins=bins, histtype='step', label='Detector C (cut)')
plt.hist(dfTarget_cut["energy"], bins=bins, histtype='step', label='Target (cut)', linewidth=2, color='black')
plt.xlabel("Energy (keV)")
plt.ylabel("Counts (after cut)")
plt.title("Energy Spectra After CNN Cut")

plt.yscale("log")


plt.legend()
plt.show()
```

```python
def model(params, pdfA, pdfB, pdfC, pdf_nldbd):

    thetaA, thetaB, thetaC, thetaNLDBD = params
    return (thetaA * pdfA +
            thetaB * pdfB +
            thetaC * pdfC +
            thetaNLDBD * pdf_nldbd)

def neg_log_likelihood(params, data_counts, pdfA, pdfB, pdfC, pdf_nldbd):

    lam = model(params, pdfA, pdfB, pdfC, pdf_nldbd)
    lam = np.clip(lam, 1e-12, None) 
    k = data_counts                               
    return np.sum(lam - k * np.log(lam))

target_counts = countsTarget_cut.astype(float)

init_guess = [1350, 770, 600, 5] 
result = minimize(
    neg_log_likelihood,
    init_guess,
    args=(target_counts, pdfA, pdfB, pdfC, nldbd_pdf),
    method='Nelder-Mead' 
)

thetaA, thetaB, thetaC, thetaNLDBD = result.x

print("Fitted values using Poisson likelihood:")
print(f"_A       = {thetaA}")
print(f"_B       = {thetaB}")
print(f"_C       = {thetaC}")
print(f"_NLDBD   = {thetaNLDBD}")

fitted_model = model(result.x, pdfA, pdfB, pdfC, nldbd_pdf)
```

```python
thetaA_best, thetaB_best, thetaC_best, thetaNLDBD_best = result.x 
NLL_min = result.fun 

def profile_nll(theta_fixed):
    initial_guess = [thetaA_best, thetaB_best, thetaC_best]
    def nll_fixed(params_free):
        thetaA, thetaB, thetaC = params_free
        params = [thetaA, thetaB, thetaC, theta_fixed]
        return neg_log_likelihood(params, target_counts, pdfA, pdfB, pdfC, nldbd_pdf)
    result_fixed = opt.minimize(nll_fixed, initial_guess, method='Nelder-Mead')
    return result_fixed.fun

delta_NLL_90CL = 2.71 / 2
NLL_threshold = NLL_min + delta_NLL_90CL
print(f"Minimum NLL = {NLL_min:.3f}. 90% CL NLL threshold = {NLL_threshold:.3f}")
def nll_diff(theta_val):
    return profile_nll(theta_val) - NLL_threshold

theta_best = thetaNLDBD_best 
if nll_diff(theta_best) > 0:
    theta_low = 0.0
else:
    theta_low = theta_best

theta_high = theta_low
while nll_diff(theta_high) < 0:
    theta_high *= 2.0 
    if theta_high == 0: 
        theta_high = 1.0 
    if theta_high > 1e5:
        raise RuntimeError("Upper limit search failed to bracket the root.")

upper_limit = opt.brentq(nll_diff, theta_low, theta_high, xtol=1e-6)
print(f"90% CL upper limit on _NLDBD  {upper_limit:.3f}")
```

```python

thetaA_bg, thetaB_bg, thetaC_bg, _ = result.x  


Asimov_counts = (
    thetaA_bg * pdfA +
    thetaB_bg * pdfB +
    thetaC_bg * pdfC
)

def neg_log_likelihood_Asimov(params, data_counts, pdfA, pdfB, pdfC, pdf_nldbd):
    """
    Poisson -Log(Likelihood) with the 'data' = Asimov_counts (no fluctuations).
    params = [_A, _B, _C, _NLDBD]
    """
    lam = model(params, pdfA, pdfB, pdfC, pdf_nldbd) 
    lam = np.clip(lam, 1e-12, None)
    k = data_counts 
    return np.sum(lam - k * np.log(lam))

init_guess = [thetaA_bg, thetaB_bg, thetaC_bg, 0.1] 
result_Asimov = minimize(
    neg_log_likelihood_Asimov,
    init_guess,
    args=(Asimov_counts, pdfA, pdfB, pdfC, nldbd_pdf),
    method='Nelder-Mead'
)

NLL_min_Asimov = result_Asimov.fun  


def profile_nll_Asimov(theta_fixed):
    """
    For a fixed _NLDBD, re-fit _A, _B, _C to get the profiled NLL
    with Asimov_counts as data.
    """
    def nll_nuisance(nuisance_params):
        tA, tB, tC = nuisance_params
        params_fixed = [tA, tB, tC, theta_fixed]
        return neg_log_likelihood_Asimov(params_fixed, Asimov_counts, pdfA, pdfB, pdfC, nldbd_pdf)
    
    init_nuisance = [result_Asimov.x[0], result_Asimov.x[1], result_Asimov.x[2]]
    res_nuisance = minimize(nll_nuisance, init_nuisance, method='Nelder-Mead')
    return res_nuisance.fun

NLL_threshold_Asimov = NLL_min_Asimov + delta_NLL_90CL

def f_Asimov(theta_val):
    return profile_nll_Asimov(theta_val) - NLL_threshold_Asimov


theta_best_Asimov = result_Asimov.x[3]
theta_low = max(0, theta_best_Asimov) 
theta_high = theta_low if theta_low > 0 else 1e-6

while f_Asimov(theta_high) < 0:
    theta_high *= 2
    if theta_high > 1e6:
        print("No sign change found up to 1e6; sensitivity is effectively infinite.")
        theta_high = 1e6
        break

thetaNLDBD_sensitivity = brentq(f_Asimov, theta_low, theta_high, xtol=1e-6)


print(f"Best-fit NLL (Asimov) = {NLL_min_Asimov:.3f}")
print(f"90% CL upper limit on _NLDBD with no true signal (Asimov) = {thetaNLDBD_sensitivity:.3f}")
```
